```python
import os
from bs4 import BeautifulSoup
import requests
```

```python
type_1_excludes_base_url = "https://www.icd10data.com/ICD10CM/Analytics/Type1_Excludes_Crosswalk"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) \
    Apple WeKit /537.36(KHTML , like Gecko) \
    Chrome/80.0.3987.162 Safari/537.36"
}
```

```python
response = requests.get(type_1_excludes_base_url, headers=headers).text
soup = BeautifulSoup(response, 'lxml')
body_content = soup.find('div', class_ = "body-content")
print(body_content)
```

```python
ul = body_content.find('ul')
li_list = ul.find_all('li', recursive=False)

results = {}

for li in li_list:
    span = li.find('span')
    if span:
        key = span.get_text(strip=True)
    values = [link.get_text(strip=True) for link in li.find_all('a', recursive=False)]
    results[key] = values
print(results)
```

```python
def extract_exculdes(url, headers=headers):    
    response = requests.get(url, headers=headers).text
    soup = BeautifulSoup(response, 'lxml')
    body_content = soup.find('div', class_ = "body-content")
    
    ul = body_content.find('ul')
    li_list = ul.find_all('li', recursive=False)

    results = {}

    for li in li_list:
        span = li.find('span')
        if not span:
            continue
        key = span.get_text(strip=True)
        values = [link.get_text(strip=True) for link in li.find_all('a', recursive=False)]
        results[key] = values
    return results

excludes_dict = {}
for i in range(1, 36):
    url = type_1_excludes_base_url + f"/{i}"
    excludes_dict.update(extract_exculdes(url))


print(excludes_dict.keys())
```

```python
import pandas as pd

df = pd.DataFrame([
    {"Code": code, "Type 1 Excludes": ", ".join(excludes)}
    for code, excludes in excludes_dict.items()
])

df
```

```python
# Save the results to a file
df.to_csv("type1_excludes.csv", index=False)
```
